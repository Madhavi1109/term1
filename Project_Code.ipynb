## 4.Data Normalisation

<a id=section401></a> 
### 4.1 Standardising STATUS

```python
source_data['Status']=source_data['Status'].map({1:"yes",2:"no"})                # Normalising the data if 1 - Yes, 2 - No
print(source_data.head())
```

<a id=section402></a> 
### 4.1 Outliers treatment

```python
for idx, feature in enumerate(list(source_data.columns)[:-1]):
    fg = sns.FacetGrid(source_data, hue='Status', size=5)
    fg.map(sns.distplot, feature).add_legend()
plt.show()
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,5))
for idx,cols in enumerate(list(source_data.columns)[:-1]):
    sns.boxplot(x='Status', y=cols, data=source_data, ax=axes[idx],showfliers=True)
plt.show()
```

<a id=section402></a> 
### Observation<br>
1.From #4.1, it can be observed that number of positive axillary nodes are higly dense from 0 to 5.<br>
2.From the above box plot, status vs pos_axillary_nodes, it can be observed that patients who are operated after 1965, the survival rate is high.<br>

## 5.Questions

#### 5.1 Has the age played any role in determining the survival status?

```python
sns.set_style('whitegrid');
sns.FacetGrid(source_data, hue = 'Status',size = 4).map(plt.scatter,'Age', 'Status').add_legend();
plt.show();
```

__Observation__
From the above plot, it is evident that if the breast cancer is detected at a young ag(i.e age<40), the survival rate is higher when somapred with age>70


#### 5.2 Based on the pos_Axillary_nodes how is the survival status

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for idx, feature in enumerate(list(source_data.columns)[:-1]):
    sns.violinplot( x='Status', y=feature, data=source_data, ax=axes[idx])
plt.show()
```

```python
sns.boxplot(x = 'Status', y = 'Pos_axillary_nodes' , data =source_data)
plt.show()
```

__Observation__
<br>
1.Survival rate is high for the people with 0 pos_axillary_nodes
